### Answer:
The data set "financials_annual" has the following problems which contradicts with the concept of tidy data:

1.) I do not have the case that every variable is placed in its own column, sine I have twenty columns for the same variable.

2.) I also do not have the case that every observation is placed in its own row, since I have observations for twenty seperate years in one row.

3.) At least the condition that every value is placed in its own cell is fulfilled.

At the end I can say that the data set financials_annual does onls fulfill one of the tidy dta conditions, which means that this data set is not in the tidy data format.
